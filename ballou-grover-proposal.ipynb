## Data Description and Acquisition

Our data will come primarily from the Utah legislative website (https://le.utah.gov/). We will scrape various aspects of bills (see Potential Variables list) from the *blank year* to 2026 General Session. The legislature website contains only public data, and, after reviewing the [robots.txt](https://le.utah.gov/robots.txt) and [terms of use](https://le.utah.gov/documents/disclaimer.htm), we are allowed to scrape the information we want. The terms of use only prohibits the use and publication of personal information/data, refering to website users with accounts/log-ins, which we do not intend to pull and are pretty sure we would not have access to scrape in the first place. The robots restricts a handful of folders from scraping (the one of note is a couple of fiscal analyst/budget information pages), but the folders that house bill information is not one of them. We also need to impose a crawl delay of 10, which we will abide by.

We will also use data from the American Community Survey (https://www.census.gov/data/datasets.html) to get our economic & demographic indicators. It is a direct download, and is widely used in research publications. We may need to clean the data to be only Utah data, but there are options to filter on the ACS website as well.

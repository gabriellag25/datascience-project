>"Do economic indicators influence the types of policy introduced?"

Our data will come primarily from the Utah legislative website (https://le.utah.gov/). ## Ethical Considerations

This project would impact Utah residents, employees of the state legislature, and potentially other state legislatures. We hope it will distill bills filled with political jargon into ordinarily comprehensible language so that normal constituents can follow, as well as encourage their civic engagement via awareness of the policy indicators discovered in our study. However, this information can be used by policymakers to inform their legislative strategy, potentially encouraging them to pass bills only in "proven successful" (via our study) subject areas and leaving other subjects undeveloped (especially for those up for reelection at the time). This information can also be used by malignant lobbying groups to improve their strategy and manipulate policymakers.

## Exploratory Analysis

Once cleaned and transformed, we plan to run summary statistics of each variable to note their means, standard deviations, min and max values, and will use a histogram to visualize their distribution and note any outliers or skews. We will then create a correlation matrix and scatterplot grid to get a preliminary understanding of strongly correlated variables. This will serve as the basis for our selection criteria for our regression and other analysis methods. Then, with our selected variables, we'll make scatterplots of each variable by session (essentially by a single year), most likely using frequency/counts of the variables over time. We want to see if any session had a jump or dip in key variables that we may want to mention or monitor. We'll also look at variable frequency by year for all dependent variables for the same reason. We may use clustering in this section to see if there were increased bill subjects in a certain session and see what kind of variables would be more influential in that year.

## Analysis Methodology

### Method 1: Linear regression

This method is to answer the questions "What aspects of a bill influence it's passage/outcome?" and "Do economic indicators influence what policy is passed?". Using the correlations we found in the previous exploratory analysis, we'll construct a regression equation with the strongest-correlated variables and do cross validation to find the best prediction model. We'll try to accurately predict the outcome of a bill, whether it is pass/fail status or final location. 

### Method 2: Classification

This method can also be used to answer the questions above, this time with a categorical variable (such as the pass/fail variable). We can also use it to determine the subject of a bill, answering the question "What influences the types of bills introduced in a session?". Decision trees will be useful here to determine relationships, as well as easy to explain to constituents which is the goal of this project. We can also use random forest for more accurate results.

We may change or add analysis methods as we learn more throughout the semester.

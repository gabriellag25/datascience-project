```python
# importing and loading packages
from bs4 import BeautifulSoup

import pandas as pd
import scipy as sc
import numpy as np

import urllib.request
import re
```

```python
# building the session list
url1 = "https://le.utah.gov/bills/billSearch.jsp"
with urllib.request.urlopen(url1) as response:
    html = response.read()
    html = html.decode('utf-8')

with open('session_list.html', 'w') as new_file:
    new_file.write(html)

sess_soup = BeautifulSoup(html, 'html.parser')

## list of sessions by full name (character string) USE THIS FOR DATA SET
sess_list = sess_soup.find(id='sess').get_text()
sess_list = sess_list.replace('Session','Session,')
sess_list = re.split(r'[,]+', sess_list)[:82]

## list of session values USE THIS FOR URLS
option = sess_soup.select("option")
val_list = []
for y in option:
    value = str(y)
    x = re.split(r'[""]+', value)
    val_list.append(x[1])

values = val_list[1:83]
```
